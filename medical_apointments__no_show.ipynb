```python
import pandas as pd
```

```python
df = pd.read_csv('/content/KaggleV2-May-2016.csv')
df.sample(50)
```

```python
df.info()
```

```python
df['AppointmentDay'] =  pd.to_datetime(df['AppointmentDay'], utc=True)
df['ScheduledDay'] =  pd.to_datetime(df['ScheduledDay'], utc=True)

df['AppointmentDay'] = df['AppointmentDay'].dt.date
df['ScheduledDay'] = df['ScheduledDay'].dt.date

df['PatientId'] = df['PatientId'].astype(int).astype(str)
df.head()
```

```python
print(df['Gender'].unique())
print(df['Scholarship'].unique())
print(df['Hipertension'].unique())
print(df['Diabetes'].unique())
print(df['Alcoholism'].unique())
print(df['Handcap'].unique())
print(df['SMS_received'].unique())
print(df['No-show'].unique())
```

```python
pd.set_option("display.max_rows", None)
df.groupby('Neighbourhood')['PatientId'].count()
```
